```python
from sklearn.model_selection import train_test_split
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_test
```

```python
y_predicted=model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test,y_test)
```
